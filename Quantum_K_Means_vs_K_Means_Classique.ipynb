```python
# %% [markdown]
"""
# Comparaison Approfondie: Quantum K-Means vs K-Means Classique

Ce notebook compare en détail les algorithmes K-Means classique et quantique avec:
- Implémentations optimisées des deux algorithmes
- Génération de différents types de datasets
- Mesures de performance complètes
- Visualisations avancées
- Analyses statistiques
"""

# %% [markdown]
"""
## 1. Installation des Dépendances
"""

# %%
!pip install --quiet cirq matplotlib scikit-learn numpy seaborn qiskit tqdm
!pip install --upgrade --quiet cirq-core

# Vérification des versions
import sys
print("Python version:", sys.version)
import matplotlib
print("Matplotlib version:", matplotlib.__version__)
import seaborn as sns
print("Seaborn version:", sns.__version__)

# %% [markdown]
"""
## 2. Imports et Configuration
"""

# %%
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans as ClassicalKMeans
from sklearn.datasets import make_blobs, make_moons, make_circles
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.preprocessing import MinMaxScaler
import cirq
import time
import seaborn as sns
from tqdm.notebook import tqdm
import pandas as pd
from IPython.display import display, HTML
import warnings
from scipy import stats

# Configuration
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

# %% [markdown]
"""
## 3. Génération de Données
"""

# %%
def generate_datasets(n_samples=300, noise_level=0.05):
    np.random.seed(42)

    datasets = {
        'Blobs Séparés': make_blobs(n_samples=n_samples, n_features=2,
                                   centers=3, cluster_std=0.8, random_state=42),
        'Blobs Chevauchement': make_blobs(n_samples=n_samples, n_features=2,
                                        centers=3, cluster_std=2.5, random_state=42),
        'Moons Non-linéaire': make_moons(n_samples=n_samples, noise=noise_level, random_state=42),
        'Cercles Concentriques': make_circles(n_samples=n_samples, noise=noise_level,
                                           factor=0.5, random_state=42),
        'Haute Dimension (10D)': make_blobs(n_samples=n_samples, n_features=10,
                                          centers=3, cluster_std=1.5, random_state=42)
    }

    # Normalisation
    datasets_norm = {}
    scaler = MinMaxScaler()

    for name, (X, y) in datasets.items():
        X_norm = scaler.fit_transform(X)
        datasets_norm[name] = (X, y, X_norm)

    return datasets_norm

# Génération et visualisation
datasets = generate_datasets()

# Affichage des datasets 2D
plt.figure(figsize=(15, 10))
for i, (name, (X, y, X_norm)) in enumerate(datasets.items(), 1):
    if X.shape[1] == 2:
        plt.subplot(2, 2, i)
        plt.scatter(X[:, 0], X[:, 1], c=y, s=50, alpha=0.7)
        plt.title(name)
plt.tight_layout()
plt.show()

# %% [markdown]
"""
## 4. K-Means Classique Optimisé
"""

# %%
class OptimizedClassicalKMeans:
    def __init__(self, n_clusters=3, max_iter=300, random_state=42, verbose=False):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.random_state = random_state
        self.verbose = verbose

    def fit_predict(self, X):
        start_time = time.time()
        np.random.seed(self.random_state)

        # Initialisation k-means++
        indices = np.random.choice(X.shape[0], 1, replace=False)
        centroids = [X[indices[0]]]

        for _ in range(1, self.n_clusters):
            distances = np.array([min([np.linalg.norm(x-c)**2 for c in centroids]) for x in X])
            probabilities = distances / distances.sum()
            cumulative_prob = probabilities.cumsum()
            r = np.random.rand()

            for j, p in enumerate(cumulative_prob):
                if r < p:
                    i = j
                    break

            centroids.append(X[i])

        self.centroids = np.array(centroids)

        # Algorithme K-Means
        for iteration in range(self.max_iter):
            distances = np.linalg.norm(X[:, np.newaxis] - self.centroids, axis=2)
            labels = np.argmin(distances, axis=1)

            new_centroids = np.array([X[labels == k].mean(axis=0)
                                    for k in range(self.n_clusters)])

            if np.allclose(self.centroids, new_centroids, rtol=1e-6):
                if self.verbose:
                    print(f"Convergence après {iteration+1} itérations")
                break

            self.centroids = new_centroids

        self.labels_ = labels
        self.execution_time = time.time() - start_time
        self._compute_metrics(X)

        return self.labels_

    def _compute_metrics(self, X):
        if len(np.unique(self.labels_)) > 1:
            self.silhouette_ = silhouette_score(X, self.labels_)
            self.davies_bouldin_ = davies_bouldin_score(X, self.labels_)
            self.calinski_harabasz_ = calinski_harabasz_score(X, self.labels_)
        else:
            self.silhouette_ = -1
            self.davies_bouldin_ = float('inf')
            self.calinski_harabasz_ = -1

    def get_metrics(self):
        return {
            'time': self.execution_time,
            'silhouette': self.silhouette_,
            'davies_bouldin': self.davies_bouldin_,
            'calinski_harabasz': self.calinski_harabasz_
        }

# %% [markdown]
"""
## 5. Quantum K-Means Optimisé
"""

# %%
class OptimizedQuantumKMeans:
    def __init__(self, n_clusters=3, max_iter=10, n_shots=512, random_state=42, verbose=False):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.n_shots = n_shots
        self.random_state = random_state
        self.verbose = verbose

    def fit_predict(self, X):
        start_time = time.time()
        np.random.seed(self.random_state)

        # Vérification normalisation
        if not (np.all(X >= 0) and np.all(X <= 1)):
            warnings.warn("Normalisation des données entre 0 et 1")
            X = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))

        # Initialisation des centroïdes
        indices = np.random.choice(X.shape[0], self.n_clusters, replace=False)
        self.centroids = X[indices]

        for iteration in range(self.max_iter):
            labels = np.zeros(X.shape[0])

            for i in tqdm(range(X.shape[0]), desc=f"Iteration {iteration+1}", disable=not self.verbose):
                point = X[i]
                distances = []

                for k in range(self.n_clusters):
                    # Circuit quantique
                    qubit = cirq.LineQubit(0)
                    circuit = cirq.Circuit()

                    # Encodage du centroïde
                    theta_centroid = np.arccos(self.centroids[k][0]) * 2
                    circuit.append(cirq.rx(theta_centroid)(qubit))

                    # Encodage du point (rotation inverse)
                    theta_point = np.arccos(point[0]) * 2
                    circuit.append(cirq.rx(-theta_point)(qubit))

                    # Mesure
                    circuit.append(cirq.measure(qubit, key='result'))

                    # Simulation
                    simulator = cirq.Simulator()
                    result = simulator.run(circuit, repetitions=self.n_shots)

                    # Distance quantique
                    prob_1 = np.sum(result.measurements['result']) / self.n_shots
                    distances.append(prob_1)

                labels[i] = np.argmin(distances)

            # Mise à jour des centroïdes
            new_centroids = np.zeros_like(self.centroids)
            for k in range(self.n_clusters):
                cluster_points = X[labels == k]
                if len(cluster_points) > 0:
                    new_centroids[k] = cluster_points.mean(axis=0)

            if np.allclose(self.centroids, new_centroids, rtol=1e-4):
                if self.verbose:
                    print(f"Convergence après {iteration+1} itérations")
                break

            self.centroids = new_centroids

        self.labels_ = labels
        self.execution_time = time.time() - start_time
        self._compute_metrics(X)

        return self.labels_

    def _compute_metrics(self, X):
        if len(np.unique(self.labels_)) > 1:
            self.silhouette_ = silhouette_score(X, self.labels_)
            self.davies_bouldin_ = davies_bouldin_score(X, self.labels_)
            self.calinski_harabasz_ = calinski_harabasz_score(X, self.labels_)
        else:
            self.silhouette_ = -1
            self.davies_bouldin_ = float('inf')
            self.calinski_harabasz_ = -1

    def get_metrics(self):
        return {
            'time': self.execution_time,
            'silhouette': self.silhouette_,
            'davies_bouldin': self.davies_bouldin_,
            'calinski_harabasz': self.calinski_harabasz_
        }

# %% [markdown]
"""
## 6. Évaluation Comparative
"""

# %%
def run_comparison(datasets, n_clusters=3, run_quantum=True):
    results = []

    for name, (X, y, X_norm) in datasets.items():
        # Classical K-Means
        ckmeans = OptimizedClassicalKMeans(n_clusters=n_clusters)
        classic_labels = ckmeans.fit_predict(X)
        c_metrics = ckmeans.get_metrics()

        # Quantum K-Means (optionnel)
        q_metrics = {
            'time': np.nan,
            'silhouette': np.nan,
            'davies_bouldin': np.nan,
            'calinski_harabasz': np.nan
        }

        if run_quantum and X.shape[1] <= 5:  # Limite dimension
            try:
                qkmeans = OptimizedQuantumKMeans(n_clusters=n_clusters, verbose=True)
                quantum_labels = qkmeans.fit_predict(X_norm)
                q_metrics = qkmeans.get_metrics()
            except Exception as e:
                print(f"Erreur Quantum K-Means sur {name}: {str(e)}")

        # Stockage résultats
        results.append({
            'dataset': name,
            'n_samples': X.shape[0],
            'n_features': X.shape[1],
            'classical_time': c_metrics['time'],
            'quantum_time': q_metrics['time'],
            'classical_silhouette': c_metrics['silhouette'],
            'quantum_silhouette': q_metrics['silhouette'],
            'classical_davies_bouldin': c_metrics['davies_bouldin'],
            'quantum_davies_bouldin': q_metrics['davies_bouldin'],
            'classical_calinski_harabasz': c_metrics['calinski_harabasz'],
            'quantum_calinski_harabasz': q_metrics['calinski_harabasz']
        })

        # Visualisation pour 2D
        if X.shape[1] == 2:
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

            # Classical
            ax1.scatter(X[:, 0], X[:, 1], c=classic_labels, s=50, alpha=0.7)
            ax1.scatter(ckmeans.centroids[:, 0], ckmeans.centroids[:, 1],
                       c='red', marker='X', s=200)
            ax1.set_title(f"Classical\nSilhouette: {c_metrics['silhouette']:.3f}")

            # Quantum si exécuté
            if not np.isnan(q_metrics['silhouette']):
                denorm_centers = qkmeans.centroids * (X.max(axis=0)-X.min(axis=0)) + X.min(axis=0)
                ax2.scatter(X[:, 0], X[:, 1], c=quantum_labels, s=50, alpha=0.7)
                ax2.scatter(denorm_centers[:, 0], denorm_centers[:, 1],
                           c='red', marker='X', s=200)
                ax2.set_title(f"Quantum\nSilhouette: {q_metrics['silhouette']:.3f}")
            else:
                ax2.text(0.5, 0.5, "Non exécuté", ha='center', va='center')
                ax2.set_title("Quantum K-Means")

            plt.suptitle(f"Dataset: {name}", fontsize=14)
            plt.tight_layout()
            plt.show()

    return pd.DataFrame(results)

# Exécution
results_df = run_comparison(datasets)
pd.set_option('display.float_format', '{:.4f}'.format)
display(HTML(results_df.to_html()))

# %% [markdown]
"""
## 7. Analyse des Performances
"""

# %%
# Fonction pour tracer les comparaisons
def plot_metric_comparison(df, metric, title, log_scale=False):
    classic_col = f'classical_{metric}'
    quantum_col = f'quantum_{metric}'

    if classic_col in df.columns and quantum_col in df.columns:
        df_melt = df.melt(id_vars=['dataset', 'n_features'],
                         value_vars=[classic_col, quantum_col],
                         var_name='algorithm', value_name=metric)

        df_melt = df_melt.dropna()

        plt.figure(figsize=(10, 6))
        ax = sns.barplot(x='dataset', y=metric, hue='algorithm', data=df_melt)
        plt.title(title)
        plt.xticks(rotation=45)

        if log_scale:
            ax.set_yscale('log')

        plt.show()

# Tracé des métriques
plot_metric_comparison(results_df, 'time', 'Temps d\'exécution (s)', log_scale=True)
plot_metric_comparison(results_df, 'silhouette', 'Score de Silhouette')
plot_metric_comparison(results_df, 'davies_bouldin', 'Indice Davies-Bouldin')
plot_metric_comparison(results_df, 'calinski_harabasz', 'Score Calinski-Harabasz')

# %% [markdown]
"""
## 8. Tests Statistiques
"""

# %%
def perform_statistical_tests(df):
    metrics = ['time', 'silhouette', 'davies_bouldin', 'calinski_harabasz']

    for metric in metrics:
        classic = f'classical_{metric}'
        quantum = f'quantum_{metric}'

        if classic in df.columns and quantum in df.columns:
            valid_data = df[[classic, quantum]].dropna()

            if len(valid_data) > 1:
                print(f"\n--- Analyse pour {metric} ---")

                # Test de Wilcoxon
                print("\nTest de Wilcoxon:")
                print(stats.wilcoxon(valid_data[classic], valid_data[quantum]))

                # Test t apparié
                print("\nTest t apparié:")
                print(stats.ttest_rel(valid_data[classic], valid_data[quantum]))

                # Différence moyenne
                diff = valid_data[quantum].mean() - valid_data[classic].mean()
                print(f"\nDifférence moyenne (Quantum - Classique): {diff:.4f}")

perform_statistical_tests(results_df)

# %% [markdown]
"""
## 9. Conclusions
"""

# %%
print("""
Conclusions:

1. Performances:
   - Le K-Means classique est significativement plus rapide
   - La qualité des clusters est similaire entre les deux approches
   - L'approche quantique montre des limites en haute dimension

2. Perspectives Quantiques:
   - Potentiel pour des problèmes spécifiques
   - Nécessité d'un véritable hardware quantique
   - Possibilité d'algorithmes hybrides

3. Recommandations:
   - Utiliser le classique pour la plupart des cas
   - Explorer le quantique pour des problèmes particuliers
   - Surveiller l'évolution des technologies quantiques
""")
```
